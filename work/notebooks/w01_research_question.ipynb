```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(df.shape)
print(df.head())
print(df.columns.tolist())
```

```python
# Example — adjust column names once you see df.columns
# Using 'sessions_last_30d' as 'traffic_after' and 'sessions_prev_30d' as 'traffic_before'
df["growth"] = df["sessions_last_30d"] - df["sessions_prev_30d"]
df["growth_pct"] = (df["growth"] / df["sessions_prev_30d"]) * 100

print(df.sort_values("growth_pct", ascending=False).head(5))   # top growing
print(df.sort_values("growth_pct", ascending=True).head(5))    # top declining
```
