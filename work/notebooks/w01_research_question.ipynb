## 3. Quick look at the data (2-3 real numbers)

```python
import pandas as pd
import numpy as np

df = pd.read_csv("content_refresh_anonymized.csv")
zero_df = df.loc[(df['impressions_90d']>=100) & (df['clicks_90d']==0)]
print('The number of pages with impressions larger than 100, but zero clicks: ',zero_df.shape[0], '\n')
print(f"That was around 20% of the dataset (of {df.shape[0]} pages)\n")
print('That statistics for each position_tier in the group: \n',zero_df['position_tier'].value_counts())
print('\n It can be seen that within the 6068 pages that have 100+ impressions, no clicks, around 20% of them are page_1 position_tier.\n')
print('This is clearly a problem, suggesting that this data is worth my next 7 weeks')
```

Pages at position_tier page_1 should be getting clicks (given they already have a fair amount of impressions).

```python
# Filter to pages with meaningful volume (100+ impressions)
substantial = df[df['impressions_90d'] >= 100]

# Group by position tier and look at CTR stats
substantial.groupby('position_tier')['ctr'].describe()
```

It is clear that the page_1 pages averages at ctr = 0.35% for 100+ impressions
-> Any page_1 pages with 100+ impressions but have significantly lower CTR are losing opportunity and are underperforming.

What can be concluded is CRT varies dramatically by position_tier (when page_1 averages at 0.35% but page_3_5 averages at 0.14%). Therefor pages at position_tier 1 that aren't getting clicks even with 100+ impressions represent opportunity.
